```python
# Imports pandas module for data analysis
import pandas as pd
```

```python
# Variable that reads in the data
reviews = pd.read_csv('ign.csv')
```

```python
# Makes new column named 'Date' that consists of year, month and day
reviews['Date'] = reviews['release_year'].map(str) +('-')+ reviews['release_month'].map(str) +('-')+ reviews['release_day'].map(str)
```

```python
# Formats column 'Date' to datetime object
reviews['Date'] = pd.to_datetime(reviews['Date'], format='%Y-%m-%d')
```

```python
# Correlation, only shows which correlations are bigger than 0.5
corr = pd.DataFrame.corr(reviews)
raw = corr[(corr.abs()>0.5)]
raw
```

```python
# Gets median score of the games
med_score = reviews.mean()['score']
med_score = round(med_score, 2)
print('Medium score is:',med_score)
```

```python
# Adds new column named 'deviation', which shows deviation between current game score and medium score of the games
reviews["deviation"]=reviews["score"] / med_score
```
